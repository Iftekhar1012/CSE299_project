```python
import os
from dotenv import load_dotenv
load_dotenv
os.environ['OPENAI_API_KEY']=os.getenv("OPENAI_API_KEY")
```

```python
from llama_index.core import VectorStoreIndex,SimpleDirectoryReader
documents=SimpleDirectoryReader("data").load_data()
```

```python
from langchain_community.document_loaders import PyPDFLoader

import os
import glob
docs = ""
loader = ""
def get_pdf_files(directory):
    pdf_files = []
    for file in glob.glob(os.path.join(directory, '*.pdf')):
        pdf_files.append(file)
    return pdf_files

# Replace 'your_folder_path' with the path to your folder containing PDF files
folder_path = 'F:\jn\data'
pdf_files = get_pdf_files(folder_path)
print(pdf_files)

for pdf_file in pdf_files:
    print("Processing:", pdf_file)
    loader=PyPDFLoader(pdf_file)+loader
    
docs=loader.load()

```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
documents=text_splitter.split_documents(docs)
documents[:5]
```

```python
## FAISS Vector Database
from langchain_community.embeddings import OllamaEmbeddings
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(documents[:20], OllamaEmbeddings())
```

```python
query = "What's the statistics about?"
retireved_results=db.similarity_search(query)
print(retireved_results[0].page_content)
```
